```python
import subprocess

# Set the path to the STAR genome index
genome_dir = "/mnt/e/new_workflow/star_index/"

# Set the path to the output directory
output_dir = "/mnt/e/new_workflow/results/4_aligned_sequences/"

# List of file names
file_names = [
    "JvRad_Rhipp_200_TSMRLT_d438483b_P0330-SB0496_S1_R1_001_trimmed.fq",
    "JvRad_Rhipp_200_TSMRLT_d438483b_P0330-SB0496_S1_R2_001_trimmed.fq",
    "JvRad_Rhipp_201_TSMRLT_a429b07e_P0330-SB0496_S2_R1_001_trimmed.fq",
    "JvRad_Rhipp_201_TSMRLT_a429b07e_P0330-SB0496_S2_R2_001_trimmed.fq",
    "JvRad_Rhipp_202_TSMRLT_b3c033d6_P0330-SB0496_S3_R1_001_trimmed.fq",
    "JvRad_Rhipp_202_TSMRLT_b3c033d6_P0330-SB0496_S3_R2_001_trimmed.fq",
    "JvRad_Rhipp_203_TSMRLT_e32edba2_P0330-SB0496_S4_R1_001_trimmed.fq",
    "JvRad_Rhipp_203_TSMRLT_e32edba2_P0330-SB0496_S4_R2_001_trimmed.fq",
    "JvRad_Rhipp_204_TSMRLT_efeebde0_P0330-SB0496_S5_R1_001_trimmed.fq",
    "JvRad_Rhipp_204_TSMRLT_efeebde0_P0330-SB0496_S5_R2_001_trimmed.fq",
    "JvRad_Rhipp_400_TSMRLT_c5c04043_P0330-SB0496_S6_R1_001_trimmed.fq",
    "JvRad_Rhipp_400_TSMRLT_c5c04043_P0330-SB0496_S6_R2_001_trimmed.fq",
    "JvRad_Rhipp_401_TSMRLT_6f156c58_P0330-SB0496_S7_R1_001_trimmed.fq",
    "JvRad_Rhipp_401_TSMRLT_6f156c58_P0330-SB0496_S7_R2_001_trimmed.fq",
    "JvRad_Rhipp_402_TSMRLT_6615aec4_P0330-SB0496_S8_R1_001_trimmed.fq",
    "JvRad_Rhipp_402_TSMRLT_6615aec4_P0330-SB0496_S8_R2_001_trimmed.fq",
    "JvRad_Rhipp_503_TSMRLT_09cb4f24_P0330-SB0496_S9_R1_001_trimmed.fq",
    "JvRad_Rhipp_503_TSMRLT_09cb4f24_P0330-SB0496_S9_R2_001_trimmed.fq",
    "JvRad_Rhipp_504_TSMRLT_cfb7f9d5_P0330-SB0496_S10_R1_001_trimmed.fq",
    "JvRad_Rhipp_504_TSMRLT_cfb7f9d5_P0330-SB0496_S10_R2_001_trimmed.fq"
]


# Run STAR for each pair of reads
for i in range(0, len(file_names), 2):
    file_R1 = file_names[i]
    file_R2 = file_names[i + 1]

    # Extract the sample name from the file name
    sample_name = file_R1.split("_R1_001_trimmed.fq")[0]

    # Run STAR
    cmd = f"STAR --genomeDir {genome_dir} --readFilesIn /mnt/e/new_workflow/results/2_trimmed_output/{file_R1} /mnt/e/new_workflow/results/2_trimmed_output/{file_R2} --runThreadN 12 --outSAMtype BAM SortedByCoordinate --quantMode GeneCounts --outFileNamePrefix {output_dir}/{sample_name}_STAR_"
    subprocess.run(cmd, shell=True)
```
